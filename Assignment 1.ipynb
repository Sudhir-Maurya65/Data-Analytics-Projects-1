```python
import matplotlib.pyplot as plt

# Step 1: Make sure Total_Sales column exists
df["Total_Sales"] = df["Quantity"] * df["Price"]

# Step 2: Group by Product Name and sum sales
product_sales = df.groupby("Product Name")["Total_Sales"].sum().reset_index()

# Step 3: Plot bar chart
plt.figure(figsize=(8,5))
plt.bar(product_sales["Product Name"], product_sales["Total_Sales"], color="pink", edgecolor="black")

plt.title("Total Sales by Product", fontsize=14)
plt.xlabel("Product Name")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Step 1: Make sure Total_Sales column exists
df["Total_Sales"] = df["Quantity"] * df["Price"]

# Step 2: Convert Purchase_Date to datetime (if not already)
df["Purchase Date"] = pd.to_datetime(df["Purchase Date"])

# Step 3: Resample data by month and sum total sales
monthly_sales = df.resample("M", on="Purchase Date")["Total_Sales"].sum().reset_index()

# Step 4: Plot line chart
plt.figure(figsize=(8,5))
plt.plot(monthly_sales["Purchase Date"], monthly_sales["Total_Sales"], marker="o", color="blue", linewidth=2)

plt.title("Monthly Sales Trend", fontsize=14)
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Step 1: Make sure Total_Sales column exists
df["Total_Sales"] = df["Quantity"] * df["Price"]

# Step 2: Plot histogram
plt.figure(figsize=(8,5))
plt.hist(df["Total_Sales"], bins=10, color="lightgreen", edgecolor="black", alpha=0.7)

plt.title("Distribution of Total Sales", fontsize=14)
plt.xlabel("Total Sales per Transaction")
plt.ylabel("Frequency")
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```
